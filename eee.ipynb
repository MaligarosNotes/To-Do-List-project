```python
# f is the no root function
def no_root_function(x):
    return x**2 + 1

def quad(a,b,c):
    # ax^2 +bx + c
    return lambda x: a*x**2 + b*x + c

def power_rule(x):
    return 2*x
    

def newtons_method(f, f_prime, x0, tol, max_iter):
    x = x0
    for i in range(max_iter):
        x = x - f(x) / f_prime(x)
        if abs(f(x)) < tol:
            return x
    return x
```
